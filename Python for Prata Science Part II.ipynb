Let's create my personal preference table

```python
userInput = [
            {'Cluster Labels':0, 'Rating':5},
            {'Cluster Labels':1, 'Rating':5},
            {'Cluster Labels':5, 'Rating':7},
            {'Cluster Labels':7, 'Rating':2},
         ] 
inputCluster = pd.DataFrame(userInput)
inputCluster
```

Based on the results of Part I, we introduce the clusters table with the different restaurant types

```python
df=pd.read_excel('Clusters.xlsx')
df=df.head(8)
df
```

```python
#Let's chjeck the cluster label type to enable the merge.
df['Cluster Labels'].dtype
```

```python
#Resetting the index to avoid future issues
df = df.reset_index(drop=True)
#Dropping unnecessary issues due to save memory and to avoid issues
userGenreTable2= df.drop('Cluster Labels', 1).head(4)
userGenreTable2
```

```python
inputCluster['Rating']
```

```python
#Dot produt to get weights
userProfile = userGenreTable2.transpose().dot(inputCluster['Rating'])
#The user profile
userProfile
```

```python
#Now let's get the genres of every restaurant in our original dataframe
genreTable = df.set_index(df['Cluster Labels'])
genreTable = genreTable.drop('Cluster Labels', 1)
#And drop the unnecessary information
genreTable
```

```python
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df
```

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
#Just a peek at the values
recommendationTable_df
```

We have the list of clusters, sorted by my hypothetical preferences, according to the recommender algorithm
